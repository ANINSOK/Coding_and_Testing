### [그래프] SilverⅡ 11724 연결 요소의 개수

### 문제
방향 없는 그래프가 주어졌을 때, 연결 요소 (Connected Component)의 개수를 구하는 프로그램을 작성하시오.

### 입력
첫째 줄에 정점의 개수 N과 간선의 개수 M이 주어진다. (1 ≤ N ≤ 1,000, 0 ≤ M ≤ N×(N-1)/2) 둘째 줄부터 M개의 줄에 간선의 양 끝점 u와 v가 주어진다. (1 ≤ u, v ≤ N, u ≠ v) 같은 간선은 한 번만 주어진다.

### 출력
첫째 줄에 연결 요소의 개수를 출력한다.

### 입출력 예


### 내 풀이

```python
def dfs(graph, visited, i):
    visited[i] = True
    for i in graph[i]:
        if not visited[i]:
            dfs(graph, visited, i)
        
N, M = map(int, input().split())
graph = [[] for _ in range(N+1)]
count = 0 # 연결 노드의 수
visited = [False] * (N+1)

for i in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

for i in range(1, N+1):
    if not visited[i]:
        dfs(graph, visited, i)
        count += 1 # dfs 한 번 끝날 때마다 count+1
print(count)
```

```python
import sys
sys.setrecursionlimit(10**6)
input = sys.stdin.readline
def dfs(visited, i, graph):
    visited[i] = True
    for i in graph[i]:
        if not visited[i]:
            dfs(visited, i, graph)

n, m = map(int, input().split())
graph = [[] for _ in range(n+1)]
cnt = 0
visited = [False] * (n+1)
for _ in range(m):
    s, e = map(int, input().split())
    graph[s].append(e)
    graph[e].append(s)

for i in range(1, n+1):
    if not visited[i]:
        dfs(visited, i, graph)
        cnt += 1
print(cnt)
```
