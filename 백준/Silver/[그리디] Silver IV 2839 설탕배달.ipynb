### [그리디] Silver IV 2839 설탕배달

### 문제 설명
상근이는 요즘 설탕공장에서 설탕을 배달하고 있다. 상근이는 지금 사탕가게에 설탕을 정확하게 N킬로그램을 배달해야 한다. 설탕공장에서 만드는 설탕은 봉지에 담겨져 있다. 봉지는 3킬로그램 봉지와 5킬로그램 봉지가 있다.

상근이는 귀찮기 때문에, 최대한 적은 봉지를 들고 가려고 한다. 예를 들어, 18킬로그램 설탕을 배달해야 할 때, 3킬로그램 봉지 6개를 가져가도 되지만, 5킬로그램 3개와 3킬로그램 1개를 배달하면, 더 적은 개수의 봉지를 배달할 수 있다.

상근이가 설탕을 정확하게 N킬로그램 배달해야 할 때, 봉지 몇 개를 가져가면 되는지 그 수를 구하는 프로그램을 작성하시오.

### 입력

첫째 줄에 N이 주어진다. (3 ≤ N ≤ 5000)

### 출력

상근이가 배달하는 봉지의 최소 개수를 출력한다. 만약, 정확하게 N킬로그램을 만들 수 없다면 -1을 출력한다. 

### 입출력 예



### 내 풀이

```python
N = int(input())

# N이 5kg로 나누어 떨어지는 경우
if N % 5 == 0:
    print(N // 5)
    
# N이 5kg로 나누어 떨어지지 않는 경우
else:
    cnt = 0  # 총 몇봉지인지 세는 변수
    
    while N > 0:    # 0kg 넘게 남아있으면,
        N -= 3      # 3kg 짜리 설탕을 하나 만들고
        cnt += 1    # 1봉지 세기

        if N % 5 == 0:      # 5로 나누어 떨어지는경우
            cnt += N // 5
            print(cnt)
            break

        elif N == 0:  # 3으로 나누어 떨어지는 경우
            print(cnt)
            break

        elif N <= 2:  # 3과 5로 모두 나누어 떨어지지 않는 경우
            print(-1)
            break
        

```

```python
N = int(input())

# N이 5의 배수인 경우
if N % 5 == 0:
    print(N // 5)
# N이 5의 배수가 아닌 경우
else:
    bag5 = N // 5  # 5kg 봉지로 최대한 채우기

    while bag5 >= 0:  # 5kg 봉지의 개수가 0 이상인 경우에 실행

        if (N - (bag5 * 5)) % 3 == 0:  # 남은 무게를 3kg 봉지로 채울 수 있는 경우
            bag3 = (N - (bag5 * 5)) // 3
            print(bag5 + bag3)
            break

        bag5 -= 1  # 5kg 봉지를 하나씩 줄여가며 시도

    else:
        print(-1)  # 5kg 봉지로는 남은 무게를 채울 수 없는 경우
```

```python
n = int(input())  # 설탕 무게 입력 받기

count = 0  # 필요한 봉지 개수를 세는 변수

while True:
    if n % 5 == 0:  # 남은 무게가 5의 배수라면
        count += n // 5  # 5kg 봉지로 모두 채우기
        print(count)  # 필요한 봉지 개수 출력
        break  # 반복 중단
    
    count += 1  # 5kg 봉지로 채우지 못하면 3kg 봉지 하나 추가
    n -= 3  # 3kg 봉지로 남은 무게 채우기
    if n < 0:  # 음수가 되면 설탕을 정확하게 채울 수 없음
        print(-1)  # -1 출력
        break  # 반복 중단
```
