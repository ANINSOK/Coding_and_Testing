### 문제
해빈이는 패션에 매우 민감해서 한번 입었던 옷들의 조합을 절대 다시 입지 않는다. 예를 들어 오늘 해빈이가 안경, 코트, 상의, 신발을 입었다면, 다음날은 바지를 추가로 입거나 안경대신 렌즈를 착용하거나 해야한다. 해빈이가 가진 의상들이 주어졌을때 과연 해빈이는 알몸이 아닌 상태로 며칠동안 밖에 돌아다닐 수 있을까?

### 입력
첫째 줄에 테스트 케이스가 주어진다. 테스트 케이스는 최대 100이다.

각 테스트 케이스의 첫째 줄에는 해빈이가 가진 의상의 수 n(0 ≤ n ≤ 30)이 주어진다.
다음 n개에는 해빈이가 가진 의상의 이름과 의상의 종류가 공백으로 구분되어 주어진다. 같은 종류의 의상은 하나만 입을 수 있다.
모든 문자열은 1이상 20이하의 알파벳 소문자로 이루어져있으며 같은 이름을 가진 의상은 존재하지 않는다.

### 출력
각 테스트 케이스에 대해 해빈이가 알몸이 아닌 상태로 의상을 입을 수 있는 경우를 출력하시오.

### 내 풀이

```python
from collections import Counter

def count_outfit_combinations(clothes):
    category_counts = Counter(category for _, category in clothes)
    
    total_combinations = 1
    for count in category_counts.values():
        total_combinations *= (count + 1)

    return total_combinations - 1

T = int(input())

for _ in range(T):
    n = int(input())
    clothes = [input().split() for _ in range(n)]

    result = count_outfit_combinations(clothes)
    print(result)
```

```python
from collections import defaultdict

# 의상 조합의 경우의 수를 계산하는 함수
def count_outfit_combinations(clothes):
    # 각 의상 종류별 의상 개수를 카운트하기 위한 딕셔너리 초기화
    outfit_counts = defaultdict(int)

    for _, category in clothes:
        # 의상 종류별로 의상 개수 카운트
        outfit_counts[category] += 1

    # 각 의상 종류별로 가능한 조합 수를 계산
    # 한 가지 의상을 추가로 입을 수 있으므로 (의상 개수 + 1)을 리스트에 저장
    combinations = [count + 1 for count in outfit_counts.values()]

    # 모든 의상 종류별 가능한 조합 수를 곱하여 전체 의상 조합 수 계산
    total_combinations = 1
    for count in combinations:
        total_combinations *= count

    # 알몸인 경우 (모든 의상을 입지 않은 경우)를 제외하기 위해 1을 뺌
    return total_combinations - 1

T = int(input())

for _ in range(T):
    n = int(input())
    clothes = [input().split() for _ in range(n)]

    result = count_outfit_combinations(clothes)
    print(result)
```

```python
from collections import defaultdict

def count_outfit_combinations(clothes):
    d = defaultdict(list)  # 의상 종류별로 의상을 그룹화하기 위한 defaultdict 생성
    def multi(arr):
        ans = 1
        for i in arr:
            ans *= (i+1)  # 주어진 리스트에 있는 값들에 1을 더한 후 모두 곱하여 반환
        return ans
    for i in range(len(clothes)):
        d[clothes[i][1]] += [clothes[i][0]]  # 의상 종류를 key로, 의상 이름을 값으로 defaultdict에 추가
    lengths = [len(v) for v in d.values()]  # 각 의상 종류별 의상 개수를 리스트에 저장
    if len(lengths) == 1:  # 의상 종류가 하나뿐이면 해당 종류의 의상 개수 반환
        return lengths[0]
    else:
        return (multi(lengths) - 1)  # 여러 의상 종류가 있을 때 가능한 조합 수 계산 후 1을 뺀 값 반환

T = int(input())

for _ in range(T):
    n = int(input())
    clothes = [input().split() for _ in range(n)]  

    result = count_outfit_combinations(clothes)
    print(result)
```
