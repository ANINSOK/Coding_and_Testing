### [구현] Silver V 11723 집합
비어있는 공집합 S가 주어졌을 때, 아래 연산을 수행하는 프로그램을 작성하시오.

- add x: S에 x를 추가한다. (1 ≤ x ≤ 20) S에 x가 이미 있는 경우에는 연산을 무시한다.

- remove x: S에서 x를 제거한다. (1 ≤ x ≤ 20) S에 x가 없는 경우에는 연산을 무시한다.

- check x: S에 x가 있으면 1을, 없으면 0을 출력한다. (1 ≤ x ≤ 20)

- toggle x: S에 x가 있으면 x를 제거하고, 없으면 x를 추가한다. (1 ≤ x ≤ 20)

- all: S를 {1, 2, ..., 20} 으로 바꾼다.

- empty: S를 공집합으로 바꾼다.


### 입력
첫째 줄에 수행해야 하는 연산의 수 M (1 ≤ M ≤ 3,000,000)이 주어진다.

둘째 줄부터 M개의 줄에 수행해야 하는 연산이 한 줄에 하나씩 주어진다.

### 출력
check 연산이 주어질때마다, 결과를 출력한다.

### 입출력 예


### 내 풀이

```python
import sys
input = sys.stdin.readline
N = int(input())
S = set()
for _ in range(N):
    parts = input().split()
    if len(parts) == 2:
        code = parts[0]
        x = int(parts[1])
    else:
        code = parts[0]
    if code == 'add':
        S.add(x)
    elif code == 'remove':
        S.discard(x)
    elif code == 'check' and x in S:
        print(1)
    elif code == 'check' and x not in S:
        print(0)
    elif code == 'toggle' and x in S:
        S.discard(x)
    elif code == 'toggle' and x not in S:
        S.add(x)
    elif code == 'all':
        S = set(range(1, 21))
    elif code == 'empty':
        S.clear()
```

```python
import sys
input = sys.stdin.readline
N = int(input())
S = set()

for _ in range(N):
    parts = input().split()
    code = parts[0]

    if code == 'add':
        x = int(parts[1])
        S.add(x)
    elif code == 'remove':
        x = int(parts[1])
        S.discard(x)
    elif code == 'check':
        x = int(parts[1])
        result = 1 if x in S else 0
        print(result)
    elif code == 'toggle':
        x = int(parts[1])
        if x in S:
            S.discard(x)
        else:
            S.add(x)
    elif code == 'all':
        S = set(range(1, 21))
    elif code == 'empty':
        S.clear()
```
