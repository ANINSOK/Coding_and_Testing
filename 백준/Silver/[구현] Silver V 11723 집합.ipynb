```python
import sys
input = sys.stdin.readline
N = int(input())
S = set()

for _ in range(N):
    parts = input().split()
    code = parts[0]

    if code == 'add':
        x = int(parts[1])
        S.add(x)
    elif code == 'remove':
        x = int(parts[1])
        S.discard(x)
    elif code == 'check':
        x = int(parts[1])
        result = 1 if x in S else 0
        print(result)
    elif code == 'toggle':
        x = int(parts[1])
        if x in S:
            S.discard(x)
        else:
            S.add(x)
    elif code == 'all':
        S = set(range(1, 21))
    elif code == 'empty':
        S.clear()
```
