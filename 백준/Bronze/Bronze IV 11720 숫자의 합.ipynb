### Bronze IV 11720. 숫자의 합

### 문제 설명
N개의 숫자가 공백 없이 쓰여있다. 이 숫자를 모두 합해서 출력하는 프로그램을 작성하시오.

### 입력

첫째 줄에 숫자의 개수 N (1 ≤ N ≤ 100)이 주어진다. 둘째 줄에 숫자 N개가 공백없이 주어진다.

### 출력

입력으로 주어진 숫자 N개의 합을 출력한다.

### 입출력 예



### 내 풀이

```python
T = int(input()) # 첫 번째 수 입력 받기
N = int(input()) # 두 번째 수 입력 받기


n = str(N)      # 입력받은 N(정수)를 str로 변환
total = 0

for i in n:     # str을 하나씩 꺼내와 int로 변환 후 계산
    total += int(i)

print(total)
```
