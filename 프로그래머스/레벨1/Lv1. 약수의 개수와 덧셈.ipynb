## LV_1. 약수의 개수와 덧셈

두 정수 left와 right가 매개변수로 주어집니다. left부터 right까지의 모든 수들 중에서, 약수의 개수가 짝수인 수는 더하고, 약수의 개수가 홀수인 수는 뺀 수를 return 하도록 solution 함수를 완성해주세요.

### 제한사항

- 1 ≤ left ≤ right ≤ 1,000

### 입출력 예

### 입출력 예 #1

- 다음 표는 13부터 17까지의 수들의 약수를 모두 나타낸 것입니다.
- 따라서, 13 + 14 + 15 - 16 + 17 = 43을 return 해야 합니다.




### 입출력 예 #2

- 다음 표는 24부터 27까지의 수들의 약수를 모두 나타낸 것입니다.
- 따라서, 24 - 25 + 26 + 27 = 52를 return 해야 합니다.



### 내 풀이

```python
left, right = 13, 17
```

```python
def solution(left, right):
    total_prime = 0

    for i in range(left, right+1):
        prime = [] # 약수 담는 바구니
        for j in range(1,i+1):
            if i % j == 0: # 약수라면 바구니에 담아라
                prime.append(j)
                
        if len(prime) % 2 == 0:  # 약수가 짝수개면 더하기
            total_prime += i
        elif len(prime) % 2 != 0: # 약수가 홀수개면 빼기
            total_prime -= i
    return total_prime

solution(left, right)
```

### 다른 사람 풀이

```python
left, right = 13, 17

def solution(left, right):
    answer = 0
    for i in range(left,right+1):
        if int(i**0.5)==i**0.5:
            answer -= i
        else:
            answer += i
    return answer

solution(left, right)
```

```python
left, right = 13, 17
import math

def solution(left, right):
    answer = 0
    for i in range(left, right + 1, 1):
        sqrt = math.sqrt(i)
        if int(sqrt) == sqrt:
            answer -= i
        else:
            answer += i

    return answer

solution(left, right)
```
