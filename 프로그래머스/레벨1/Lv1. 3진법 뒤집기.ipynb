## LV_1. 3진법 뒤집기

자연수 n이 매개변수로 주어집니다. n을 3진법 상에서 앞뒤로 뒤집은 후, 이를 다시 10진법으로 표현한 수를 return 하도록 solution 함수를 완성해주세요.

### 제한사항

- n은 1 이상 100,000,000 이하인 자연수입니다.

### 내 풀이

```python
n = 125
```

```python
def solution(n):
    answer = ''

    while n > 0:
        n, rest = divmod(n, 3) # divmod : 몫(n)과 나머지(rest)를 반환해주는 함수
        answer += str(rest)  # answer에 n의 나머지를 문자로 변환하여 차곡차곡 저장

    answer = int(answer, 3)  # str(answer)를 3진수로 변환해줘!

    return answer
solution(n)
```

### 다른 사람 풀이

```python
n = 125

def solution(n):
    tmp = ''
    while n != 0:
        tmp += str(n % 3)  #divmod 없이 몫과 나머지를 하나씩 직접 구해 풀이
        print(tmp)
        n = n // 3
        print(n)

    answer = int(tmp, 3)
    return answer

solution(n)
```
